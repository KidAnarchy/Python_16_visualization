```python
from matplotlib import rcParams

rcParams.update({'font.size':12})

x = np.arange(0, 10, 0.05)
y1 = np.sin(5*x)
y2 = np.sin(6*x)
y = y1 + y2
ybeat = 2*np.cos(0.5*(6-5)*x)

# plot 1st wave
plt.xlabel('time(ms)')
plt.ylabel('Signal')

plt.ylim(-2, 4)   # range of y
#plt.scatter(x,y1)
plt.plot(x, y1, 'gs', label='1st wave')  # 'g'green colour  's'regtanger
# label='1st wave' is legend
plt.legend()
plt.show()

# Plot 2nd wave
plt.xlabel('time(ms)')
plt.ylabel('Signal')

plt.ylim(-2, 4)   # range of y
#plt.scatter(x,y1)
plt.plot(x, y2, 'ro', label='2st wave')  # 'r'red colour  'o'oval
# label='1st wave' is legend
plt.legend()
plt.show()
```

```python
from matplotlib import rcParams

rcParams.update({'font.size':12})

x = np.arange(0, 10, 0.05)
y1 = np.sin(5*x)
y2 = np.sin(6*x)
y = y1 + y2
ybeat = 2*np.cos(0.5*(6-5)*x)

# plot 1st wave
plt.xlabel('time(ms)')
plt.ylabel('Signal')

plt.ylim(-2, 4)   # range of y
#plt.scatter(x,y1)
plt.plot(x, y1, 'gs', label='1st wave')  # 'g'green colour  's'regtanger
# label='1st wave' is legend
plt.legend()
#plt.show()  #ถ้าไม่ต้องการแยกกราฟให้ลบ

# Plot 2nd wave
plt.xlabel('time(ms)')
plt.ylabel('Signal')

plt.ylim(-2, 4)   # range of y
#plt.scatter(x,y1)
plt.plot(x, y2, 'ro', label='2st wave')  # 'r'red colour  'o'oval
# label='2st wave' is legend
plt.legend()
plt.show()

# 1st+2st wave
plt.xlabel('time(ms)')
plt.ylabel('Signal')
plt.ylim(-2, 4)   # range of y
plt.plot(x, y, 'k-',linewidth=5, label='1st+2st wave')  # 'k' Black colour  '-' line
plt.plot(x, ybeat, 'b--',linewidth=2, label='beats')
# linewidth = ความหนาของเส้น
plt.legend()
plt.show()
```
